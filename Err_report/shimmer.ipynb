```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_au_intensities(csv_file, title):
    # CSV 파일 읽기
    df = pd.read_csv(csv_file)
    
    # 타임스탬프 컬럼이 있을 경우 사용, 없으면 인덱스를 사용
    if 'timestamp' in df.columns:
        time = df['timestamp']
    else:
        time = df.index
        
    # '_r'로 끝나는 컬럼 추출 (Intensity 값)
    au_columns = [col for col in df.columns if col.endswith('_r')]
    
    plt.figure(figsize=(12, 8))
    for au in au_columns:
        plt.plot(time, df[au], label=au)
    
    plt.ylim(0, 5)  # y축 범위 설정 (0 ~ 5)
    plt.xlabel('Time (ms or frame index)')
    plt.ylabel('Intensity')
    plt.title(title)
    plt.legend()
    plt.show()

# 본인 녹화 비디오 (Changmook.csv) 그래프 출력
plot_au_intensities('/Users/changmookoh/Downloads/Changmook.csv', 'Action Unit Intensities - Self Recorded Video')

# 과제 비디오 (44.csv) 그래프 출력
plot_au_intensities('/Users/changmookoh/Downloads/44.csv', 'Action Unit Intensities - Video 44')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
file_path = "/Users/changmookoh/Downloads/yechan_shimmer_data_session1_2_20250316_0929PM.csv"
df = pd.read_csv(file_path)
# CSV 파일 불러오기 (이미 df 변수에 로드된 상태)
ppg_signal = df['heart_rate_raw'].values

# 1. 신호 시각화
plt.figure(figsize=(10, 4))
plt.plot(ppg_signal)
plt.title("Raw PPG Signal (heart_rate_raw)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()

# 2. 센서 타임스탬프 차이를 이용해 샘플링 주파수 추정
time_diffs = df['sensor_timestamp'].diff().dropna().values
print("타임스탬프 차이(일부):", time_diffs[:10])

# 2. 센서 타임스탬프 차이를 이용해 샘플링 주파수 추정
# 만약 sensor_timestamp의 단위가 틱(tick)이고, 1 tick = 1/8192 초라면:
tick_to_seconds = 1.0 / 8192.0
time_interval = np.median(time_diffs) * tick_to_seconds  # 초 단위
fs = 1.0 / time_interval
print("추정된 샘플링 주파수: {:.2f} Hz".format(fs))


# 3. 피크 검출 (피크 간 최소 간격은 예를 들어 0.5초 정도로 설정)
peaks, _ = find_peaks(ppg_signal, distance=fs*0.5)
print("검출된 피크 개수:", len(peaks))

# 검출된 피크를 신호에 표시해보기
plt.figure(figsize=(10, 4))
plt.plot(ppg_signal, label='PPG Signal')
plt.plot(peaks, ppg_signal[peaks], "x", label='Detected Peaks')
plt.legend()
plt.title("PPG Signal with Detected Peaks")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()

# 4. 인터비트 간격(IBI) 계산 (초 단위)
ibi = np.diff(peaks) / fs
print("평균 IBI (초):", np.mean(ibi))

# 5. 심박수 계산 (bpm)
hr = 60.0 / np.mean(ibi)
print("계산된 심박수: {:.2f} bpm".format(hr))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# CSV 파일 불러오기
file_path = "/Users/changmookoh/Downloads/yechan_shimmer_data_session1_2_20250316_0929PM.csv"
df = pd.read_csv(file_path)

# PPG 신호로 사용 (heart_rate_raw가 PPG 신호라고 가정)
ppg_signal = df['heart_rate_raw'].values

# 샘플링 주파수 계산: 
# 주어진 sensor_timestamp 간격은 256 tick으로 일정하고, 1 tick = 1/8192초 (0.00012207초)라면,
tick_to_sec = 1.0 / 8192.0  
# 각 샘플 간 시간 = 256 * tick_to_sec = 256/8192 = 0.03125초 -> fs = 1/0.03125 = 32 Hz
fs = 32.0  
print("샘플링 주파수: {:.2f} Hz".format(fs))

# 슬라이딩 윈도우 설정 (8초 = 8*32 = 256 샘플, non-overlapping)
window_length = int(8 * fs)  # 256 samples
step = window_length  # non-overlapping (원하는 경우 step을 조절하세요)

# 윈도우별 심박수 계산
hr_values = []      # 각 윈도우의 심박수 (bpm)
window_timestamps = []  # 윈도우 중심의 sensor_timestamp (원하는 경우 시간 변환 가능)

for start in range(0, len(ppg_signal) - window_length + 1, step):
    window_signal = ppg_signal[start:start+window_length]
    
    # 피크 검출: 최소 피크 간격은 약 0.5초 (fs*0.5 샘플)
    peaks, _ = find_peaks(window_signal, distance=fs*0.5)
    
    # 피크가 2개 이상 검출된 경우에만 IBI 계산
    if len(peaks) > 1:
        ibi = np.diff(peaks) / fs  # 초 단위 IBI
        median_ibi = np.median(ibi)  # 중앙값 IBI (노이즈에 강함)
        hr_window = 60.0 / median_ibi  # bpm 계산
    else:
        hr_window = np.nan  # 피크가 충분치 않으면 NaN 처리
    
    hr_values.append(hr_window)
    # 윈도우 중심의 timestamp 저장 (여기서는 원시 sensor_timestamp 사용)
    window_center = df['sensor_timestamp'].iloc[start + window_length//2]
    window_timestamps.append(window_center)

# 결과를 새로운 DataFrame으로 생성
hr_df = pd.DataFrame({
    "window_center_timestamp": window_timestamps,
    "heart_rate_calculated": hr_values
})

print(hr_df)

# 전체 평균 심박수 (유효한 값만 평균)
mean_hr = np.nanmean(hr_values)
print("전체 평균 심박수: {:.2f} bpm".format(mean_hr))

# 결과 시각화
plt.figure(figsize=(10,4))
plt.plot(hr_df["window_center_timestamp"], hr_df["heart_rate_calculated"], marker="o")
plt.xlabel("Sensor Timestamp (tick)")
plt.ylabel("Calculated HR (bpm)")
plt.title("Sliding Window HR Calculation")
plt.show()
```

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy.signal import find_peaks

# CSV 파일들이 들어있는 폴더 경로 설정 (예시 경로)
folder_path = "/Users/changmookoh/Downloads/pyshimmer_data"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
print(f"총 {len(csv_files)}개의 CSV 파일이 발견되었습니다.")

# Shimmer 장치: 
# sensor_timestamp 간격이 256 tick, 1 tick = 1/8192초로 가정 → 각 샘플 간 시간 = 256/8192 = 0.03125초, 즉 fs = 32 Hz
fs = 32.0
window_length = int(8 * fs)  # 8초 = 256 샘플
step = window_length         # 겹치지 않음

# 각 CSV 파일마다 처리
for file_path in csv_files:
    print("\n처리중:", file_path)
    df = pd.read_csv(file_path)
    
    # PPG 신호 (heart_rate_raw)를 사용한다고 가정
    ppg_signal = df['heart_rate_raw'].values
    
    # system_timestamp는 Unix timestamp(초 단위)라고 가정하고,
    # 이를 UTC로 로컬라이즈한 후 'America/Los_Angeles' (PST)로 변환합니다.
    df['datetime'] = pd.to_datetime(df['system_timestamp'], unit='s')\
                        .dt.tz_localize('UTC').dt.tz_convert('America/Los_Angeles')
    
    hr_list = []       # 각 윈도우별 심박수 (bpm)
    time_list = []     # 각 윈도우 중심의 실제 datetime
    
    # 슬라이딩 윈도우별 심박수 계산
    for start in range(0, len(ppg_signal) - window_length + 1, step):
        window_signal = ppg_signal[start:start+window_length]
        
        # 피크 검출: 최소 피크 간격 약 0.5초 (fs*0.5 샘플)
        peaks, _ = find_peaks(window_signal, distance=fs*0.5)
        
        if len(peaks) > 1:
            ibi = np.diff(peaks) / fs  # 피크 간격 (초)
            median_ibi = np.median(ibi)
            hr_window = 60.0 / median_ibi  # bpm 계산
        else:
            hr_window = np.nan  # 피크가 충분하지 않으면 NaN 처리
        
        hr_list.append(hr_window)
        center_idx = start + window_length // 2
        time_list.append(df['datetime'].iloc[center_idx])
    
    # 파일별 결과 DataFrame 생성
    hr_df = pd.DataFrame({
        "datetime": time_list,
        "heart_rate_calculated": hr_list
    })
    
    # 분 단위 그룹화: 각 datetime을 분 단위로 내림 (floor('T'))
    hr_df['minute'] = hr_df['datetime'].dt.floor('T')
    
    # 각 분별 통계 (최소, 최대, 평균, 표준편차)
    group = hr_df.groupby('minute')['heart_rate_calculated']\
                 .agg(min_hr='min', max_hr='max', mean_hr='mean', std_hr='std')\
                 .reset_index()
    
    # 결과 테이블 출력
    print(f"\n[{os.path.basename(file_path)}] 분별 심박수 통계:")
    print(group)
    
    # 분별 심박수 범위를 에러바로 나타내는 그래프 그리기
    plt.figure(figsize=(12, 6))
    # 에러바: 중앙은 평균, 아래쪽 에러 = (mean - min), 위쪽 에러 = (max - mean)
    plt.errorbar(group['minute'], group['mean_hr'], 
                 yerr=[group['mean_hr'] - group['min_hr'], group['max_hr'] - group['mean_hr']],
                 fmt='o', ecolor='blue', capsize=5, markersize=4, linestyle='-', linewidth=1.5, label='HR Range')
    plt.xlabel("Time (PST)")
    plt.ylabel("Heart Rate (bpm)")
    plt.title(f"Minute-wise Heart Rate for {os.path.basename(file_path)}")
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.gcf().autofmt_xdate()  # x축 라벨 자동 회전
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
