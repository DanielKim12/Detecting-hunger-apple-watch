```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_au_intensities(csv_file, title):
    # CSV  
    df = pd.read_csv(csv_file)
    
    #     ,   
    if 'timestamp' in df.columns:
        time = df['timestamp']
    else:
        time = df.index
        
    # '_r'    (Intensity )
    au_columns = [col for col in df.columns if col.endswith('_r')]
    
    plt.figure(figsize=(12, 8))
    for au in au_columns:
        plt.plot(time, df[au], label=au)
    
    plt.ylim(0, 5)  # y   (0 ~ 5)
    plt.xlabel('Time (ms or frame index)')
    plt.ylabel('Intensity')
    plt.title(title)
    plt.legend()
    plt.show()

#    (Changmook.csv)  
plot_au_intensities('/Users/changmookoh/Downloads/Changmook.csv', 'Action Unit Intensities - Self Recorded Video')

#   (44.csv)  
plot_au_intensities('/Users/changmookoh/Downloads/44.csv', 'Action Unit Intensities - Video 44')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
file_path = "/Users/changmookoh/Downloads/yechan_shimmer_data_session1_2_20250316_0929PM.csv"
df = pd.read_csv(file_path)
# CSV   ( df   )
ppg_signal = df['heart_rate_raw'].values

# 1.  
plt.figure(figsize=(10, 4))
plt.plot(ppg_signal)
plt.title("Raw PPG Signal (heart_rate_raw)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()

# 2.       
time_diffs = df['sensor_timestamp'].diff().dropna().values
print(" ():", time_diffs[:10])

# 2.       
#  sensor_timestamp  (tick), 1 tick = 1/8192 :
tick_to_seconds = 1.0 / 8192.0
time_interval = np.median(time_diffs) * tick_to_seconds  #  
fs = 1.0 / time_interval
print("  : {:.2f} Hz".format(fs))


# 3.   (      0.5  )
peaks, _ = find_peaks(ppg_signal, distance=fs*0.5)
print("  :", len(peaks))

#    
plt.figure(figsize=(10, 4))
plt.plot(ppg_signal, label='PPG Signal')
plt.plot(peaks, ppg_signal[peaks], "x", label='Detected Peaks')
plt.legend()
plt.title("PPG Signal with Detected Peaks")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()

# 4.  (IBI)  ( )
ibi = np.diff(peaks) / fs
print(" IBI ():", np.mean(ibi))

# 5.   (bpm)
hr = 60.0 / np.mean(ibi)
print(" : {:.2f} bpm".format(hr))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# CSV  
file_path = "/Users/changmookoh/Downloads/yechan_shimmer_data_session1_2_20250316_0929PM.csv"
df = pd.read_csv(file_path)

# PPG   (heart_rate_raw PPG  )
ppg_signal = df['heart_rate_raw'].values

#   : 
#  sensor_timestamp  256 tick , 1 tick = 1/8192 (0.00012207),
tick_to_sec = 1.0 / 8192.0  
#     = 256 * tick_to_sec = 256/8192 = 0.03125 -> fs = 1/0.03125 = 32 Hz
fs = 32.0  
print(" : {:.2f} Hz".format(fs))

#    (8 = 8*32 = 256 , non-overlapping)
window_length = int(8 * fs)  # 256 samples
step = window_length  # non-overlapping (  step )

#   
hr_values = []      #    (bpm)
window_timestamps = []  #   sensor_timestamp (    )

for start in range(0, len(ppg_signal) - window_length + 1, step):
    window_signal = ppg_signal[start:start+window_length]
    
    #  :     0.5 (fs*0.5 )
    peaks, _ = find_peaks(window_signal, distance=fs*0.5)
    
    #  2    IBI 
    if len(peaks) > 1:
        ibi = np.diff(peaks) / fs  #   IBI
        median_ibi = np.median(ibi)  #  IBI ( )
        hr_window = 60.0 / median_ibi  # bpm 
    else:
        hr_window = np.nan  #    NaN 
    
    hr_values.append(hr_window)
    #   timestamp  (  sensor_timestamp )
    window_center = df['sensor_timestamp'].iloc[start + window_length//2]
    window_timestamps.append(window_center)

#   DataFrame 
hr_df = pd.DataFrame({
    "window_center_timestamp": window_timestamps,
    "heart_rate_calculated": hr_values
})

print(hr_df)

#    (  )
mean_hr = np.nanmean(hr_values)
print("  : {:.2f} bpm".format(mean_hr))

#  
plt.figure(figsize=(10,4))
plt.plot(hr_df["window_center_timestamp"], hr_df["heart_rate_calculated"], marker="o")
plt.xlabel("Sensor Timestamp (tick)")
plt.ylabel("Calculated HR (bpm)")
plt.title("Sliding Window HR Calculation")
plt.show()
```

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy.signal import find_peaks

# CSV      ( )
folder_path = "/Users/changmookoh/Downloads/pyshimmer_data"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
print(f" {len(csv_files)} CSV  .")

# Shimmer : 
# sensor_timestamp  256 tick, 1 tick = 1/8192       = 256/8192 = 0.03125,  fs = 32 Hz
fs = 32.0
window_length = int(8 * fs)  # 8 = 256 
step = window_length         #  

#  CSV  
for file_path in csv_files:
    print("\n:", file_path)
    df = pd.read_csv(file_path)
    
    # PPG  (heart_rate_raw)  
    ppg_signal = df['heart_rate_raw'].values
    
    # system_timestamp Unix timestamp( ) ,
    #  UTC   'America/Los_Angeles' (PST) .
    df['datetime'] = pd.to_datetime(df['system_timestamp'], unit='s')\
                        .dt.tz_localize('UTC').dt.tz_convert('America/Los_Angeles')
    
    hr_list = []       #    (bpm)
    time_list = []     #     datetime
    
    #    
    for start in range(0, len(ppg_signal) - window_length + 1, step):
        window_signal = ppg_signal[start:start+window_length]
        
        #  :     0.5 (fs*0.5 )
        peaks, _ = find_peaks(window_signal, distance=fs*0.5)
        
        if len(peaks) > 1:
            ibi = np.diff(peaks) / fs  #   ()
            median_ibi = np.median(ibi)
            hr_window = 60.0 / median_ibi  # bpm 
        else:
            hr_window = np.nan  #    NaN 
        
        hr_list.append(hr_window)
        center_idx = start + window_length // 2
        time_list.append(df['datetime'].iloc[center_idx])
    
    #   DataFrame 
    hr_df = pd.DataFrame({
        "datetime": time_list,
        "heart_rate_calculated": hr_list
    })
    
    #   :  datetime    (floor('T'))
    hr_df['minute'] = hr_df['datetime'].dt.floor('T')
    
    #    (, , , )
    group = hr_df.groupby('minute')['heart_rate_calculated']\
                 .agg(min_hr='min', max_hr='max', mean_hr='mean', std_hr='std')\
                 .reset_index()
    
    #   
    print(f"\n[{os.path.basename(file_path)}]   :")
    print(group)
    
    #       
    plt.figure(figsize=(12, 6))
    # :  ,   = (mean - min),   = (max - mean)
    plt.errorbar(group['minute'], group['mean_hr'], 
                 yerr=[group['mean_hr'] - group['min_hr'], group['max_hr'] - group['mean_hr']],
                 fmt='o', ecolor='blue', capsize=5, markersize=4, linestyle='-', linewidth=1.5, label='HR Range')
    plt.xlabel("Time (PST)")
    plt.ylabel("Heart Rate (bpm)")
    plt.title(f"Minute-wise Heart Rate for {os.path.basename(file_path)}")
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.gcf().autofmt_xdate()  # x   
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
